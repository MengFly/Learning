# 两信号相加和相乘 

```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
input_x = np.linspace(-2*math.pi, 2*math.pi, 1000)
```

```python
out_sin_1 = np.sin(input_x)
```

```python
plt.plot(input_x, out_sin_1, 'r')
```

```python
out_sin_8 = np.sin(8*input_x)
```

```python
plt.plot(input_x, out_sin_8, 'g')
plt.plot(input_x, out_sin_1, "r")
```

```python
sum_out = out_sin_1 + out_sin_8
```

```python
plt.plot(input_x,sum_out)
plt.plot(input_x, out_sin_1+1, 'y')
plt.plot(input_x, out_sin_1-1, 'y')
```

```python
mul_out = out_sin_1 * out_sin_8
```

```python
plt.plot(input_x, mul_out)
plt.plot(input_x, out_sin_1, 'y')
plt.plot(input_x, -out_sin_1, 'y')
```
